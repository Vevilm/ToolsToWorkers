# [dbo].[ToolsView]

```sql
/****** Object:  View [dbo].[ToolsView]    Script Date: 20.10.2024 0:30:43 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[ToolsView] AS
SELECT
Tools.ID AS ID,
Articles.Name AS Name,
Storages.Name AS Storage,
Tools.Status AS Status,
Articles.ID AS Article,
Articles.Description AS Description,
Articles.Weight AS Weight,
StorageID,
Storages.Type AS Type,
Storages.Status AS StorageStatus
FROM Tools INNER JOIN Storages ON Tools.StorageID = Storages.ID
INNER JOIN Articles ON Tools.ArticleID = Articles.ID
GO
```
